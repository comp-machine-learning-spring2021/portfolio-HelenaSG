# Data Visualization

#### Importing packages 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
import string
import nltk
import nltk.corpus
nltk.download('punkt')
nltk.download('stopwords')
from nltk.probability import FreqDist
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
from wordcloud import WordCloud
from wordcloud import STOPWORDS
from collections import Counter
```

#### Our data

```python
df = pd.read_csv("text_data.csv")
df.head()
```

#### Creating word clouds

Concatenate all positive/negative texts as two strings:

```python
df_pos = df[(df['sentiment']=='pos')]
text_lst_pos = df_pos['text'].values.tolist()

text_pos = ''
for t in text_lst_pos:
  text_pos = text_pos + str(t) + ' '
```

```python
df_neg = df[(df['sentiment']=='neg')]
text_lst_neg = df_neg['text'].values.tolist()

text_neg = ''
for t in text_lst_neg:
  text_neg = text_neg + str(t) + ' '
```

Manually filter out high frequency words that are either neutral or too general:

```python
stop_words = ['Kronos', 'government', 'GAStech', 'PUBLISHED', 'LOCATION', 'ABILA', 
              'will', 'POK', 'one', 'said', 'people', 'new', 'time', 'group', 'today', 
              'report', 'year', 'report', 'work'] + list(STOPWORDS)
```

```python
# The negative cloud
wordcloud_neg = WordCloud(stopwords = stop_words, width = 800, height = 800, background_color ='white').generate(text_neg)                 
plt.figure(figsize = (8, 8))
plt.imshow(wordcloud_neg)
plt.axis("off")
plt.show()
```

```python
# The positive cloud
wordcloud_pos = WordCloud(stopwords = stop_words, width = 800, height = 800, background_color ='white').generate(text_pos)                 
plt.figure(figsize = (8, 8))
plt.imshow(wordcloud_pos)
plt.axis("off")
plt.show()
```
